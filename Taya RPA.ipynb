```python
# Preparing the environment
# install selenium lib for web automation with python using ´$ pip install selenium´
from selenium import webdriver 
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select

TAYA_WEBSITE = 'https://imobiliario.desenvolvimento.taya.com.br';

EMAIL_INPUT = '//*[@id=":r0:"]';
PASSWORD_INPUT = '//*[@id=":r1:"]';
LOGIN_BUTTON = '//*[@id="root"]/div[1]/div/div/div[2]/div/div/div/div/div/div/form/div[4]/button';

SIMULATE_BUTTON = '//*[@id="root"]/div[1]/div[1]/div/div[2]/ul/div[3]/div/span';

PROPERTY_ACQUISITION_BUTTON = '//*[@id="root"]/div[1]/div/div/div[2]/div[2]/div[1]';
MOVE_FOWARD_BUTTON = '//*[@id="root"]/div[1]/div/div/div[3]/div/div';

PARTNER_INPUT_OPTIONS = '//*[@id=":r6:"]';
CPF_INPUT = '//*[@id=":r8:"]';
NAME_INPUT = '//*[@id=":rb:"]';
BIRTH_DATE_INPUT = '//*[@id=":rc:"]';
INCOME_INPUT = '//*[@id=":rf:"]';
PHONE_INPUT = '//*[@id=":ri:"]';
USER_EMAIL_INPUT = '//*[@id=":rl:"]';
CONTINUE_BUTTON = '//*[@id="root"]/div[1]/div/div/div/div[2]/div/div/form/div/div[8]/div/button';

RESIDENTIAL_BUTTON = '//*[@id="root"]/div[1]/div/div/div/div[2]/div/div/form/div/div[1]/div/div/label[1]';
PROPERTY_UF = '//*[@id=":rr:"]';
PROPERTY_VALUE_INPUT = '//*[@id=":rs:"]'
FINANCING_VALUE_INPUT = '//*[@id=":rt:"]'
FINANCING_TERM_INPUT = '//*[@id=":r10:"]';
CALCULATE_BUTTON = '//*[@id="root"]/div[1]/div/div/div/div[2]/div/div/form/div/div[7]/div/button'


USER_EMAIL = 'heitorflorido@gmail.com';
USER_PASSWORD = '1234567890';
USER_PARTNER = 'AAI CONTROLE LTDA';
USER_CPF = '52998224725'; # random cpf number
USER_NAME = 'TEST TEST TEST';
USER_BIRTH_DATE = '11112000';
USER_INCOME = '11111';
USER_PHONE = '21123456789';
PROPERTY_VALUE = '200000'
FINANCING_VALUE = '100000'
FINANCING_TERM = '40';
USER_UF = 'AC'

# Method to wait loading page and avoid errors by not finding some element
def wait_loading(xpath):
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, xpath)));
 
# initialize an instance of the chrome driver (browser)
driver = webdriver.Chrome();

# visit taya site
driver.get(TAYA_WEBSITE);
```

```python
# Login page
# Wait page loading
wait_loading(EMAIL_INPUT);

# Fill login form
driver.find_element('xpath', EMAIL_INPUT).send_keys(USER_EMAIL);
driver.find_element('xpath', PASSWORD_INPUT).send_keys(USER_PASSWORD);

# Click on login button
driver.find_element('xpath', LOGIN_BUTTON).click();
```

```python
# Simulate page
# Wait page loading
wait_loading(SIMULATE_BUTTON);

# CLick on simulate button on the sidebar
driver.find_element('xpath', SIMULATE_BUTTON).click();
```

```python
# Product page
# Click on 'AQUISIÇÃO DE IMÓVEL' option and on button 'AVANÇAR'
wait_loading(PROPERTY_ACQUISITION_BUTTON);

driver.find_element('xpath', PROPERTY_ACQUISITION_BUTTON).click();
driver.find_element('xpath', MOVE_FOWARD_BUTTON).click();
```

```python
# Client page
wait_loading(PARTNER_INPUT_OPTIONS);

# Select partner using dropdown
driver.find_element('xpath', PARTNER_INPUT_OPTIONS).send_keys(USER_PARTNER)
driver.find_element('xpath', "//*[contains(text(), 'AAI CONTROLE LTDA')]").click();

# Fill user data fields
driver.find_element('xpath', CPF_INPUT).send_keys(USER_CPF);
driver.find_element('xpath', NAME_INPUT).send_keys(USER_NAME);
driver.find_element('xpath', BIRTH_DATE_INPUT).send_keys(USER_BIRTH_DATE);
driver.find_element('xpath', INCOME_INPUT).send_keys(USER_INCOME);
driver.find_element('xpath', PHONE_INPUT).send_keys(USER_PHONE);
driver.find_element('xpath', USER_EMAIL_INPUT).send_keys(USER_EMAIL);

# Click on 'CONTINUAR' button
driver.find_element('xpath', CONTINUE_BUTTON).click();
```

```python
# Condition page
wait_loading(RESIDENTIAL_BUTTON)

driver.find_element('xpath', RESIDENTIAL_BUTTON).click();

# Select property uf using dropdown
driver.find_element('xpath', PROPERTY_UF).click();
driver.find_element('xpath', f"//*[contains(text(),{USER_UF})]").click();

# Fill property data fields
driver.find_element('xpath', PROPERTY_VALUE_INPUT).send_keys(PROPERTY_VALUE);
driver.find_element('xpath', FINANCING_VALUE_INPUT).send_keys(FINANCING_VALUE);
driver.find_element('xpath', FINANCING_TERM_INPUT).send_keys(FINANCING_TERM);

# Click on 'CALCULAR' button
driver.find_element('xpath', CALCULATE_BUTTON).click();
```
